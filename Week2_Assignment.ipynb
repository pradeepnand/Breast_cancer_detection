### Step 7: Create a correlation matrix and mention strongly, weakly and negatively correlated quantities. 

```python
df.iloc[:,1:].corr(method="pearson")
```

#### Correlation values ranges from -1 to 1, below are the strong,week and negitive quantities.

zero means no correlationship between the variables.

1 or -1 means perfectly correlated.

As the value get closer to 1 or -1 means stonger positive or negitive correlation between the variables.

0.5 or -0.5 means moderate correlated.

Between 0.5 to < 0 or -0.5 to > 0 would have a weak positive or negitive correlationship, As it get closed to 0 weeker the correlation.

### why we need to Drop features having high correlation:

The reason why highly correlated variables (Multicollinearity) are avoided when creating models is because they can skew the output. If there are two independent variables that are representing the same occurrence it can create “noise” or inaccuracy in the model,(overfitting) where the model may be great on known training set, but will fail at unknown testing set. As this leads to higher standard error with lower statistical significance value, multicollinearity makes it difficult to ascertain how important a feature is to the target variable.
